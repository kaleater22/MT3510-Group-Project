FUNCTION synchronizing_pair_graph(A = (Q, Σ, τ, qs, F)):

    -- SETUP
    -- Create all possible pairs of states {q1, q2}, including singletons {q, q}
    -- Mark all singletons as "visited" and add them to the queue

    -- BUILD REVERSE EDGES
    -- For every pair {q1, q2} and every letter a:
    --     work out where q1 and q2 get sent by a
    --     add a reverse edge: that result <-- {q1, q2}

    -- BFS
    -- While the queue is not empty:
    --     take a pair off the queue
    --     for each pair that has a reverse edge pointing to it:
    --         if we haven't visited it yet:
    --             mark it visited, add to queue

    -- CHECK
    -- If every non-singleton pair was visited: return TRUE (synchronizing)
    -- Otherwise: return FALSE (not synchronizing)

```python
#Part C - Pair graph

def synchronizing_pair_graph(A):
    
    Q, Sigma, tau, qs, F = A

    # Each pair is stored as a frozenset so {q1,q2} == {q2,q1}
    # {q,q} then automatically becomes a singleton

    all_pairs = set()
    for q1 in Q:
        for q2 in Q:
            all_pairs.add(frozenset({q1, q2}))

    singletons = set()
    for node in all_pairs:
        if len(node) == 1:
            singletons.add(node)

    # For each pair and each letter find the image
    # Store the reverse: result <-- pair

    reverse_edges = {node: set() for node in all_pairs}
    for node in all_pairs:
        node_list = list(node)
        if len(node_list) == 1:
            q1, q2 = node_list[0], node_list[0]
        else:
            q1, q2 = node_list[0], node_list[1]
        for a in Sigma:
            target = frozenset({tau[(q1, a)], tau[(q2, a)]})
            reverse_edges[target].add(node)

    # Flood backwards from singletons — any pair reached can be merged

    visited = set(singletons)
    queue = list(singletons)

    while queue:
        current = queue.pop(0)
        for neighbour in reverse_edges[current]:
            if neighbour not in visited:
                visited.add(neighbour)
                queue.append(neighbour)

    # Check every non-singleton pair was reached
    # If so, every pair can be merged and a reset word exists

    non_singletons = all_pairs - singletons

    if non_singletons.issubset(visited):
        return True
    else:
        return False
```
